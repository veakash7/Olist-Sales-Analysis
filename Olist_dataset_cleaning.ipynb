# E-Commerce Analysis

```python
import pandas as pd
import numpy as np
orders=pd.read_csv('/content/drive/MyDrive/Data Analysis/E-Commerce Sales/olist_orders_dataset.csv')
order_items=pd.read_csv('/content/drive/MyDrive/Data Analysis/E-Commerce Sales/olist_order_items_dataset.csv')
order_payments=pd.read_csv('/content/drive/MyDrive/Data Analysis/E-Commerce Sales/olist_order_payments_dataset.csv')
order_reviews=pd.read_csv('/content/drive/MyDrive/Data Analysis/E-Commerce Sales/olist_order_reviews_dataset.csv')
products=pd.read_csv('/content/drive/MyDrive/Data Analysis/E-Commerce Sales/olist_products_dataset.csv')
sellers=pd.read_csv('/content/drive/MyDrive/Data Analysis/E-Commerce Sales/olist_sellers_dataset.csv')
customers=pd.read_csv('/content/drive/MyDrive/Data Analysis/E-Commerce Sales/olist_customers_dataset.csv')
product_category=pd.read_csv('/content/drive/MyDrive/Data Analysis/E-Commerce Sales/product_category_name_translation.csv')
```

```python
orders.info()
orders.isna().sum()
```

```python
orders['delivery_status']=orders['order_delivered_customer_date'].notnull().map({True:'Delivered',False:'Not Delivered'})
orders.head(5)
```

```python
orders.info()
fix_cols=['order_delivered_carrier_date','order_delivered_customer_date','order_estimated_delivery_date','order_purchase_timestamp','order_approved_at']
for cols in fix_cols:
  orders[cols]=pd.to_datetime(orders[cols],errors='coerce')
```

```python
order_reviews.info()
order_reviews.isna().sum()
```

```python
order_reviews.info()
fix_col=['review_creation_date','review_answer_timestamp']
for col in fix_col:
  order_reviews[col]=pd.to_datetime(order_reviews[col],errors='coerce')
order_reviews.dtypes
```

```python
orders['delivery_date_missing'] = (
    (orders['order_status'] == 'delivered') & (orders['order_delivered_customer_date'].isna())
)
```

```python
import os
output_dir = r'A:\Data Analysis\E-Commerce Sales\cleaned datasets'
os.makedirs(output_dir, exist_ok=True)
```

```python
customers.to_csv(os.path.join(output_dir, "customers_cleaned.csv"), index=False)
orders.to_csv(os.path.join(output_dir, "orders_cleaned.csv"), index=False)
order_items.to_csv(os.path.join(output_dir, "order_items_cleaned.csv"), index=False)
order_payments.to_csv(os.path.join(output_dir, "order_payments_cleaned.csv"), index=False)
order_reviews.to_csv(os.path.join(output_dir, "order_reviews_cleaned.csv"), index=False)
products.to_csv(os.path.join(output_dir, "products_cleaned.csv"), index=False)
sellers.to_csv(os.path.join(output_dir, "sellers_cleaned.csv"), index=False)
product_category.to_csv(os.path.join(output_dir, "category_translation_cleaned.csv"), index=False)
```
